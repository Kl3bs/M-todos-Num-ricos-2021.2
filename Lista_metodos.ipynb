```python
def bisection(a, b, error):
    #Iterations
    n = (np.log(b-a) - np.log(error)/np.log(2))
    n = np.ceil(n)
    index = 0
     
    while index < n:
        if f(a)*f(b) > 0:
            print("Is not possible to confirm if there is any root in this interval")
        else:
            m = round((a+b)/2)
            if f(a) * f(m) < 0:
                b = m
            else:
                a = m
                
        print(f'x{index+1} = {m}')
        index+=1
```

```python
bisection(0,100, 1)
```
